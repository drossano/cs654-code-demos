```python
def get_X(P):
    r = []
    for x in P:
        r.append(x[0])
    return r
def get_Y(P):
    r = []
    for x in P:
        r.append(x[1])
    return r
```

```python
X = get_X(points)
Y = get_Y(points)
```

```python
import matplotlib.pyplot as plot
%matplotlib inline
plot.ylim([0, 105])
plot.xlim([0,105])
X = get_X(points)
Y = get_Y(points)
plot.scatter(X, Y)
```
