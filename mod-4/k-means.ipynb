```python
points = [[52, 28],[25, 95], [36, 64],[20, 94],[52, 27],[20, 96],[39, 65],[34, 90],[75, 15],
[74, 21],
[90, 43],
[60, 12],
[67, 20],
[30, 92],
[61, 31],
[8, 74],
[51, 32],
[64, 41], [15, 80], [9, 80]]
```

```python
def get_X(P):
    r = []
    for x in P:
        r.append(x[0])
    return r
def get_Y(P):
    r = []
    for x in P:
        r.append(x[1])
    return r
```

```python
import matplotlib.pyplot as plot
%matplotlib inline
plot.ylim([0, 105])
plot.xlim([0,105])
X = get_X(points)
Y = get_Y(points)
plot.scatter(X, Y)
```

```python
import random
centroid_1 = random.choice(points)
centroid_2 = random.choice(points)
while centroid_2==centroid_1:
    centroid_2= random.choice(points)
print("1:", centroid_1, "2: ", centroid_2)
```

```python
plot.ylim([0, 105])
plot.xlim([0,105])
X = get_X(points)
Y = get_Y(points)
plot.scatter(X, Y)
X = get_X([centroid_1])
Y = get_Y([centroid_1])
plot.ylim([0, 105])
plot.xlim([0,105])
plot.scatter(X, Y, color="green", marker="^")
X = get_X([centroid_2])
Y = get_Y([centroid_2])
plot.ylim([0, 105])
plot.xlim([0,105])
plot.scatter(X, Y, color="red", marker="^")
```

```python
import math

def distance(p,q):
    return int(math.fabs(p[0] - q[0]) + math.fabs(p[1]-q[1]))
```

```python
cluster_1 = []
cluster_2 = []
```

```python
for point in points:
    if(distance(point, centroid_1) < distance(point, centroid_2)):
        cluster_1.append(point)
    else:
        cluster_2.append(point)
```

```python
cluster_points = []
for x in cluster_1:
    cluster_points.append(x)

X=get_X(cluster_points)
Y=get_Y(cluster_points)

plot.ylim([0, 105])
plot.xlim([0,105])
plot.scatter(X, Y, color = "green")

cluster_points = []
for x in cluster_2:
    cluster_points.append(x)

X=get_X(cluster_points)
Y=get_Y(cluster_points)

plot.ylim([0, 105])
plot.xlim([0,105])
plot.scatter(X, Y, color = "red")
```

```python
def SSE(center,cluster):
    sum = 0
    for p in cluster:
        d = distance(center, p)
        sum += + d * d
    return sum
```

```python
sse_1 = SSE(centroid_1, cluster_1)
sse_2 = SSE(centroid_2, cluster_2)   
print(sse_1, ", ", sse_2)
```

```python
def new_centroid(cluster):
    sum_x = 0
    sum_y = 0
    for point in cluster:
        sum_x += point[0]
        sum_y += point[1]
    return int(sum_x/len(cluster)), int(sum_y/len(cluster))
        
```

```python
old_centroid_1 = centroid_1
old_centroid_2 = centroid_2

centroid_1 = new_centroid(cluster_1)
centroid_2 = new_centroid(cluster_2)
print("1: ", old_centroid_1, " ", centroid_1, "\n2: ", old_centroid_2, " ", centroid_2)
```
