```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

points = [[8,9],[4,6],[7,5],[5,2],[2,7],[8,4]]
```

```python
def get_X(P):
    r = []
    for x in P:
        r.append(x[0])
    return r
def get_Y(P):
    r = []
    for x in P:
        r.append(x[1])
    return r
```

```python
X = get_X(points)
Y = get_Y(points)
```

```python
import matplotlib.pyplot as plot
%matplotlib inline

X = get_X(points)
Y = get_Y(points)
plot.scatter(X, Y)
```

```python
labels = []
for i in range(len(points)):
    labels.append(str(i))
```

```python

%matplotlib inline

x = get_X(points)
y = get_Y(points)
plot.scatter(X, Y, s = 100)
for label, x, y in zip(labels, x, y):
    plot.annotate(label, xy = (x,y), xytext=(5,5), textcoords='offset points', fontsize = 8)
```

```python
# min distance
z = linkage(points, method ='single')

plot.figure(figsize=(10,7))
dendrogram(z)
plot.show()
```

```python
# max distance
z = linkage(points, method ='complete')

plot.figure(figsize=(10,7))
dendrogram(z)
plot.show()
```

```python
# avg distance
z = linkage(points, method ='average')

plot.figure(figsize=(10,7))
dendrogram(z)
plot.show()
```
